# Ipython magic

Special command from ipython which simplify life

* `%autoreload` from `%load_ext autoreload` which allow to reload module without restarting kernel
* `%gui` - enable integration with populat GUI backend
* `%pdb`, `%debug` - debuging support
* `%lsmagic`- List currently available magic functions.
* `%matplotlib` - control plotting integration
* `%time` - measure execution of statement
* `%timeit` - measure execution of statement with repetition



Full list https://ipython.readthedocs.io/en/stable/interactive/magics.html

## Base plotting

````python
import numpy as np
from matplotlib import pyplot as plt

points = np.linspace(start=-7, stop=7, num=100)
plt.figure(figsize=(6,2.5), dpi=150)
plt.plot(points, np.sin(points), label="sin(x)");
````

## Addling labels 

````python
import numpy as np
from matplotlib import pyplot as plt

points = np.linspace(start=-7, stop=7, num=100)
plt.figure(figsize=(6,2.5), dpi=150)
plt.plot(points, np.sin(points), label="sin(x)")
plt.plot(points, np.cos(points), label="cos(x)")
plt.legend();
````

## Styling

````python
import numpy as np
from matplotlib import pyplot as plt

points = np.linspace(start=-7, stop=7, num=100)
with plt.style.context("ggplot"):
    plt.figure(figsize=(6,2.5), dpi=150)
    plt.plot(points, np.sin(points), label="sin(x)", linestyle="-.")
    plt.plot(points, np.cos(points), label="cos(x)", linestyle=':')
    plt.legend();
````

# Reload

Using ``%autoreload`` magic allow to modify code without need to reload data and pasting whole code to notebook. 
This reload is done before execute of each cell.
But it have limitations:

1.   New method added to class will not shown in already created object instances
2.   For complex cases reload it may take long time
3.   After all user should verify if all code work

````python
Documentation https://ipython.readthedocs.io/en/stable/config/extensions/autoreload.html
````
